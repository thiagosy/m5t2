```python
# importação dos módulos

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
import numpy as np
import pdfplumber
import re
```

```python
# leitura do arquivo csv com pandas

dados = pd.read_csv('b1abb33c4f.csv')
```

```python
# transformação do csv em DataFrame

df = pd.DataFrame(dados)
```

```python
# conversão dos títulos em minúsculo (desnecessário aqui, mas melhor para manipular)

df.columns = df.columns.str.lower()

df.head()
df['peso']
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# plotando 

plt.figure(figsize=(3,5))
sns.boxplot(data=df, y='idademae')
plt.title('Idade da mãe')
plt.ylabel('Idades')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# plotando 

plt.figure(figsize=(3,5))
sns.boxplot(data=df, y='peso')
plt.title('Peso dos bebês (em gramas)')
plt.ylabel('Peso')
plt.show()
```
